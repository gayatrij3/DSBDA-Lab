```python
iris = load_iris()
iris.keys()
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df["label"] = iris.target
iris.target_names
```

```python

iris_df.shape
```

```python

iris_df.head()
```

```python

iris_df.info()
```

```python

iris_df.describe()
```

```python
setosa = iris_df[iris_df["label"] == 0].drop("label", axis=1)
setosa.describe()
```

```python
plt.figure(figsize=(10,7))
ax = setosa.boxplot()
medians = setosa.median()
for i, col in enumerate(setosa.columns):
    ax.annotate(
        f"{medians[col]:.2f}",
        (i + 1, medians[col]),
        ha="center",
        va="center",
        color="red",
        size=10
    )
plt.show()
```
