```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
```

```python
plt.scatter(titanic["age"], titanic["fare"])
```

```python
x = np.arange(1, 11)
y_pos = x * 2
y_neg = -x * 2
y_none = np.random.randint(1, 20, size=10)
plt.figure(figsize=(10,5))
plt.scatter(x, y_pos, color='green', label='Positive Correlation')
plt.scatter(x, y_neg, color='red', label='Negative Correlation')
plt.scatter(x, y_none, color='blue', label='No Correlation')
plt.title("Scatter Plot: Positive, Negative and No Correlation")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

```python
plt.scatter(titanic["age"], titanic["fare"], 
            s=titanic["pclass"]*60, alpha=0.5)
plt.title("Bubble Chart: Age vs Fare vs Passenger Class")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
values = [63.8,-8.1,-23.2,-14.1,11.0,10.9,2.9,1.5,-3.8,-0.7,2.5,7.3]
cumulative = np.cumsum([0] + values[:-1])
colors = ["green" if v >= 0 else "red" for v in values]
plt.figure(figsize=(10,5))
plt.bar(labels, values, bottom=cumulative, color=colors)
net = sum(values)
plt.bar("Net", net, color="skyblue")
for i, v in enumerate(values):
    plt.text(i, cumulative[i] + v/2, str(v), ha='center', color="black")
plt.text(len(labels), net/2, str(round(net,1)), ha='center')
plt.axhline(0, color='black', linewidth=0.8)
plt.title("Waterfall Chart")
plt.ylabel("Values")

plt.show()
```

```python
x = np.random.normal(50,10,100)
y = np.random.normal(30,5,100)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, x+y)
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")
ax.set_zlabel("Distribution")
plt.title("3D Distribution Plot")
plt.show()
```

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
x, y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.title("3D Area Chart")
plt.show()
```
