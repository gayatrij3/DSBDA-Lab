1.Load Datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
df = pd.read_csv('titanic.csv')
```

2.Boxtplot

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['Age'].dropna())
plt.title("Box Plot - Age")
plt.show()
```

3. Pie Chart

```python
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Survival Distribution")
plt.ylabel("")
plt.show()
```

4. Scatter Plot

```python
plt.scatter(df['Age'], df['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

5. Treemap

```python
sizes = df['Pclass'].value_counts()
labels = sizes.index.astype(str)
squarify.plot(sizes=sizes, label=labels)
plt.axis('off')
plt.show()
```

6. Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

7. Waterfall Chart

```python
labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
values = [63.8,-8.1,-23.2,-14.1,11.0,10.9,2.9,1.5,-3.8,-0.7,2.5,7.3]
cumulative = np.cumsum([0] + values[:-1])
colors = ["green" if v >= 0 else "red" for v in values]
plt.figure(figsize=(10,5))
plt.bar(labels, values, bottom=cumulative, color=colors)
net = sum(values)
plt.bar("Net", net, color="skyblue")
for i, v in enumerate(values):
    plt.text(i, cumulative[i] + v/2, str(v), ha='center', color="black")
plt.text(len(labels), net/2, str(round(net,1)), ha='center')
plt.axhline(0, color='black', linewidth=0.8)
plt.title("Waterfall Chart")
plt.ylabel("Values")

plt.show()
```

8. Bubble Chart

```python
plt.scatter(df['Age'], df['Fare'], s=df['Fare'], alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Bubble Chart")
plt.show()
```

9. Histogram

```python
plt.hist(df['Age'].dropna(), bins=20)
plt.title("Age Distribution")
plt.show()
```

10. Line Plot

```python
df.groupby('Pclass')['Fare'].mean().plot()
plt.title("Average Fare by Class")
plt.show()
```
